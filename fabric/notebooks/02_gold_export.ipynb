# Gold layer: aggregate and export CSV
This notebook reads processed Parquet from the silver layer, performs aggregations or business logic, and writes the result as CSV files (as requested). Adjust grouping/aggregation logic to your needs.

```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col
spark = SparkSession.builder.getOrCreate()
print('Spark session ready:', spark)
```

```python
# Read silver parquet
silver_path = '<ONE_LAKE_ROOT>/fabric/silver/<table_name>'
df = spark.read.parquet(silver_path)
print('Silver rows:', df.count())
display(df.limit(10))
```

```python
# Example aggregation - replace with real business logic
# If you have a timestamp and category columns, aggregate counts per day/category
if len(df.columns) >= 2:
    key = df.columns[0]
    df_gold = df.groupBy(key).count()
else:
    df_gold = df.limit(100)
print('Gold rows (preview):', df_gold.count())
display(df_gold.limit(20))
```

```python
# Write gold as CSV files to gold path (update gold_path variable)
gold_path = '<ONE_LAKE_ROOT>/fabric/gold/<table_name>_csv'
df_gold.repartition(1).write.mode('overwrite').option('header', 'true').csv(gold_path)
print('Wrote gold CSV to', gold_path)
```
